```python
import json
import matplotlib.pyplot as plt

# Load the data from the JSON files
with open('./data/server/unobfuscated.json') as f:
    unobfuscated_data = json.load(f)

with open('./data/server/tigress.json') as f:
    tigress_data = json.load(f)

for unobf in unobfuscated_data: 
    if unobf["category"] != "miners": 
        continue 
    
    transformations = [] 
    file_sizes = []
    for obf in tigress_data: 
        if obf["name"] != unobf["name"]:
            continue 
        
        transformations.append(obf["transformation"])
        file_sizes.append(obf["file_size"])
        
    plt.axhline(y=unobf["file_size"], color='gray', linestyle='--', zorder=-1)
    plt.bar(transformations, file_sizes)
    plt.ylabel("Hash rate")
    plt.xticks(rotation=45, ha='right')
    plt.title(unobf["name"])
    plt.savefig(f'./plots/tigress/file-size/{unobf["name"]}.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Load the data from the JSON files
with open('./data/server/unobfuscated.json') as f:
    unobfuscated_data = json.load(f)

with open('./data/server/wasm-mutate.json') as f:
    wasm_mutate_data = json.load(f)

for unobf in unobfuscated_data: 
    if unobf["category"] != "miners": 
        continue 
    
    iterations = [] 
    file_sizes = []
    for obf in wasm_mutate_data: 
        if obf["name"] != unobf["name"]:
          continue 
            
        if obf.get("hash_rate") == None: 
          continue
        
        iterations.append(obf["iteration"] / 100)
        file_sizes.append(obf["file_size"])

    plt.axhline(y=unobf["file_size"], color='gray', linestyle='--', zorder=-1)
    plt.bar(iterations, file_sizes)
    plt.ylabel("File size")
    
    plt.xticks(iterations, [str(int(i * 100)) for i in iterations], rotation=45, ha='right')  # Set xticks labels
    plt.title(unobf["name"])
    plt.savefig(f'./plots/wasm-mutate/file-size/{unobf["name"]}.png', dpi=300, bbox_inches='tight')
    plt.show()
```
