## File Size

```python
import json
import matplotlib.pyplot as plt

# Load the JSON data
with open("../../data/file_size.json", "r") as f:
    file_sizes = json.load(f)

# Iterate over the JSON objects and plot the graphs
for item in file_sizes:
    algorithm = item["algorithm"]
    original_file_size = item[f"{algorithm}.wasm"]
    file_names = []
    relative_sizes = []
    
    for file_name, file_size in item.items():
        if file_name != "algorithm" and file_name != f"{algorithm}.wasm":
            file_names.append(file_name.split('-')[-1].split('.wasm')[0])
            relative_sizes.append(file_size / original_file_size)

    # Plot the graph
    plt.figure(figsize=(10, 5))
    plt.axhline(y=1.0, color='gray', linestyle='--', zorder=0)  # Add a gray dashed line at 1.0, behind the bars
    plt.bar(file_names, relative_sizes, zorder=1)
    plt.xticks(rotation=45, ha="right")
    plt.title(f"Relative File Sizes for {algorithm}")
    plt.ylabel("Relative File Size")
    plt.xlabel("Transformation")
    plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Load the JSON data
with open("../../data/hash_rates.json", "r") as f:
    file_sizes = json.load(f)

# Iterate over the JSON objects and plot the graphs
for item in file_sizes:
    algorithm = item["algorithm"]
    original_hash_rate = item[f"{algorithm}"]
    file_names = []
    relative_hash_rates = []
    
    for file_name, hash_rate in item.items():
        if file_name != "algorithm" and file_name != f"{algorithm}":
            file_names.append(file_name.split('-')[-1].split('.wasm')[0])
            relative_hash_rates.append(hash_rate / original_hash_rate)

    # Plot the graph
    plt.figure(figsize=(10, 5))
    plt.axhline(y=1.0, color='gray', linestyle='--', zorder=0)  # Add a gray dashed line at 1.0, behind the bars
    plt.bar(file_names, relative_hash_rates, zorder=1)  # Changed variable name here
    plt.xticks(rotation=45, ha="right")
    plt.title(f"Relative Hash Rates for {algorithm}")
    plt.ylabel("Relative Hash Rate")
    plt.xlabel("Transformation")
    plt.show()
```
